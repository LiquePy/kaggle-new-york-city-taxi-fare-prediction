```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.metrics import mean_squared_error
#import os
#print(os.listdir('../input/nyc-taxi-fare-data-preparation/test_ready_to_model.csv'))
Nrows = 2_000_000

train_data = pd.read_csv('../input/nyc-taxi-fare-initial-split/stack_training_data.csv', nrows=Nrows)
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(train_data, test_size=0.2, random_state=1)
train_data_labels = train_data.fare_amount
test_data_labels = test_data.fare_amount

train_data.drop('fare_amount', axis=1, inplace=True)
test_data.drop('fare_amount', axis=1, inplace=True)

from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor(n_estimators=150)
forest_reg.fit(train_data, train_data_labels)


# predict
prediction_data =  pd.read_csv('../input/nyc-taxi-fare-initial-split/stack_validation_data.csv', nrows=Nrows)

final_predictions_csv = forest_reg.predict(prediction_data.values)

submission = pd.DataFrame(data={'fare_amount': final_predictions_csv})
submission.to_csv('submission.csv', index=None)
measured_data =  pd.read_csv('../input/nyc-taxi-fare-initial-split/stack_validation_label.csv', nrows=Nrows, header=None)

test_csv = pd.read_csv('../input/nyc-taxi-fare-data-preparation/test_ready_to_model.csv')
test_csv_key = test_csv.key.values
test_csv.drop('key', axis=1, inplace=True)
final_predictions_csv_new = forest_reg.predict(test_csv.values)
submission_of_final_test = pd.DataFrame(data={'key': test_csv_key, 'fare_amount': final_predictions_csv_new})
submission_of_final_test.to_csv('submission_of_final_test.csv', index=None)
submission_of_final_test.head()
```
